```python
import pandas as pd
```

```python
data=pd.read_excel("En_In.xls", skiprows=17)
data=data.iloc[0:, 2:]
data = data.rename(columns={"Unnamed: 2": "Country", "Petajoules": "Energy Supply", "Gigajoules":"Energy Supply per Capita","%":"% Renewable"})

data.head()
```

```python
data.replace('...',None, inplace=True)
data['Energy Supply'] =data['Energy Supply'] * 1000000
data.head()
```

```python
country_replacements = {
    "Republic of Korea": "South Korea",
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"
}
data['Country'] = data['Country'].replace(country_replacements)
data.head()
```

```python
data['Country'] = data['Country'].str.replace(r"\s*\(.*\)|\d+", "", regex=True)
data.loc[data["Country"].isin(["American Samoa","South Korea","Bolivia"])]
```

```python
gdp = pd.read_csv('gpd.csv', skiprows=4)


gdp_replacements = {
    "Korea, Rep.": "South Korea",
    "Iran, Islamic Rep.": "Iran",
    "Hong Kong SAR, China": "Hong Kong"
}
gdp['Country Name'] = gdp['Country Name'].replace(gdp_replacements)

columns_to_keep = ['Country Name',"Country Code","Indicator Name","Indicator Code"] + [str(year) for year in range(2006, 2016)]
gdp = gdp[columns_to_keep]


gdp.head(1)
```

```python
scima=pd.read_excel("scimagojr.xlsx")
scima.head()
```

```python
data_merged = pd.merge(scima, data, how='left', left_on='Country', right_on='Country')
data_merged = pd.merge(data_merged, gdp, how='left', left_on='Country', right_on='Country Name')

columns_to_keep = [
    'Country','Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document',
    'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable'
] + [str(year) for year in range(2006, 2016)]

data_merged = data_merged[columns_to_keep]
data_merged.set_index("Country",inplace=True)
data_merged=data_merged.sort_values(by="Rank")
data_merged=data_merged.head(15)
print(data_merged.shape)
data_merged.head()
```

```python
def task8():
    res = data_merged[[str(year) for year in range(2006, 2016)]].mean(axis=1).sort_values(ascending=False)
    return res

res = task8()
print(res)
```

```python
def task9():
    avg = task8()
    country = avg.index[4]  # 5-а країна (0-based indexing)
    gdp_diff = data_merged.loc[country, '2015'] - data_merged.loc[country, '2006']
    return country, gdp_diff

print(task9())
```

```python
def task10():
    max_country = data_merged['% Renewable'].idxmax()
    max_value = data_merged.loc[max_country, '% Renewable']
    return max_country, max_value

print(task10())
```

```python
def task11():
    data_merged['People'] = data_merged['Energy Supply'] / data_merged['Energy Supply per Capita']
    data_sorted = data_merged['People'].sort_values(ascending=False)
    country = data_sorted.index[5]
    population = data_sorted.iloc[5]
    return country, population

print(task11())
```

```python
def task12():
    data_merged['Citations per Capita'] = data_merged['Citations'] / data_merged['People']
    correlation_value = data_merged[['Citations per Capita', 'Energy Supply per Capita']].corr()
    return correlation_value.iloc[0,1]

print(task12())
```

```python
def task13():
    median_value = data_merged['% Renewable'].median()
    data_merged['Renewable Flag'] = (data_merged['% Renewable'] >= median_value).astype(int)
    return data_merged['Renewable Flag']

print(task13())
```

```python
ContinentDict = {
        'China': 'Asia', 'United States': 'North America', 'Japan': 'Asia',
        'United Kingdom': 'Europe', 'Russian Federation': 'Europe', 'Canada': 'North America',
        'Germany': 'Europe', 'India': 'Asia', 'France': 'Europe', 'South Korea': 'Asia',
        'Italy': 'Europe', 'Spain': 'Europe', 'Iran': 'Asia', 'Australia': 'Australia',
        'Brazil': 'South America'
    }
def task14():
    

    data_merged['Continent'] = data_merged.index.to_series().map(ContinentDict)
    population_stats = data_merged.groupby('Continent')['People'].agg(['size', 'sum', 'mean', 'std'])
    return pd.DataFrame(population_stats)

task14()
```

```python
import matplotlib.pyplot as plt

def task15():
    data_merged['Continent'] = data_merged.index.to_series().map(ContinentDict)
    plt.figure(figsize=(10, 6))

    scatter = plt.scatter(
        data_merged['Rank'], data_merged['% Renewable'],
        s=data_merged['2015'] / 1e10,  # Масштабування ВВП
        c=data_merged['Continent'].map({'Asia': 'red', 'Europe': 'blue', 'North America': 'green',
                                        'South America': 'yellow', 'Australia': 'purple'}),
        alpha=0.6, edgecolors='w', linewidth=0.5
    )
    plt.xlabel('Rank')
    plt.ylabel('% Renewable')
    plt.title('Renewable Energy vs Rank (Bubble Size ~ GDP 2015)')
    plt.show()

task15()
```
